```python
import cv2
from numpy import *
import readtraces as rt
import os
import sys
from matplotlib import pyplot as pl
from matplotlib import cm
from PIL import Image
%matplotlib inline
```

```python
os.chdir('/media/corinna/corinna-data/data/20150107_Many_droplet_system_height_test_6mm_wide_15wtpcTTAB_5uL_50um_Droplets_2x_Olympus_4fps/')
mov=rt.movie('0p5mm_high_2_output.avi')
f=mov.getFrame(0)
r=cv2.HoughCircles(f[:,:,2],cv2.cv.CV_HOUGH_GRADIENT,3,2000,minRadius=450,maxRadius=600).astype(int).flatten()
print r
b=f.copy()
cv2.circle(b,(r[0],r[1]),r[2]-10, (0,0,255),5)
pl.imshow(b)
mask=zeros(mov.shape[::-1])
cv2.circle(mask,(r[0],r[1]),r[2]-10,1,-1)
```

```python
f=mov.getFrame(3000)
vorim=f.copy()
contim=f.copy()
f=f[:,:,2]
fcopy=cv2.GaussianBlur(f,(51,51),0)
fcopy=fcopy*mask+255*(1-mask)
print amax(fcopy), amin(fcopy)
thresh=(fcopy<180).astype(uint8)

print amax(thresh), amin(thresh)
blobs,contours=rt.extract_blobs(thresh, -1, (10,3000), -1, diskfit=False,returnCont=True, outpSpac=1)
# thresh=dstack((thresh,thresh,thresh))
cv2.drawContours(contim,contours,-1,(255,0,0),2)
pl.imshow(contim)
Image.fromarray(contim).save('/home/corinna/python/swimmertracking/contim.png')
vor=rt.Voronoi(blobs[:,3:5])
for i in range(blobs.shape[0]):
    r=vor.regions[vor.point_region[i]]
    col=tuple([int(255*c) for c in cm.jet(i*255/len(vor.points))])[:3]
#     pl.plot(vor.vertices[r,1],vor.vertices[r,0], c=col[:3])
    cv2.polylines(vorim, [(vor.vertices[r]).astype(int32)], True, col[:3], 2)
Image.fromarray(vorim).save('/home/corinna/python/swimmertracking/vorim.png')
print vorim.shape
```
